Как можно заметить, поиск в глубину, поиск в ширину и алгоритм Дейкстры ведут себя почти одинаково на произвольных графах без весов. Алгоритм Беллмана — Форда же тратит очень много времени.

```python
fig = plt.figure()
plt.title("Тест на рандомных данных c неотрицательными весами")
plt.xlabel("Количество вершин")
plt.ylabel("Время в секундах")
x = [100,200,300,400,500]
bf_y = [0.2753554,2.072154,6.8734297,17.6597932,35.266672]
plt.scatter(x,bf_y,color=(0,0,0))
dijkstra_y = [0.0029929,0.009977,0.023944,0.0478878,0.0778179]
plt.scatter(x,dijkstra_y,color=(0,1,0))
```

У алгоритма Дейкстры время растёт очень медленно, в сравнии с алгоритмом Беллмана - Форда. Попробуем проверить результаты 4 алгоритмов на цепи.

```python
fig = plt.figure()
plt.title("Тест на цепи фиксированной длины")
plt.xlabel("Количество вершин")
plt.ylabel("Время в секундах")
x = [1000,2000,3000,4000,5000]
bfs_y = [0.0009977,0.0019956,0.002993,0.0034962,0.0049885]
plt.scatter(x,bfs_y,color=(0,0,1),s=20)
dfs_y = [0.0009975,0.0019955,0.0029949,0.0039905,0.0039906]
plt.scatter(x,dfs_y,color=(1,0,0),s=20)
dijkstra_y = [0.0180062,0.0688397,0.1560746,0.2936567,0.4534841]
plt.scatter(x,dijkstra_y,color=(0,1,0),s=20)
bf_y = [0.7172701,2.8667962,6.6551056,13.5305261,22.6777789]
plt.scatter(x,dijkstra_y,color=(0,1,0),s=20)
```

Как и на первом графике все алгоритма ведут себя одинаково, кроме алгоритма Беллмана — Форда. Таким образом, если дан граф без весов, то стоит использовать любой алгоритм кроме алгоритма Беллмана — Форда. Если дан граф с неотрицательными весами, то нужно использовать Дейкстру, а если также присутствуют отрицательные веса, то алгоритма Беллмана — Форда.
